```python
import pandas as pd
s=pd.Series([1,2,3,4,5])
print('To Multiply all values in a series by 2')
print('---------------------------------------------')
print(s*2)
print('To Find Square of all values in a series')
print('---------------------------------------------')
print(s**2)
print('to print all values that are greater than 2')
print('---------------------------------------------')
print(s[s>2])
```

```python
import pandas as pd
s1=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
s2=pd.Series([10,20,30,40,50],index=['a','b','c','d','e'])
s3=pd.Series([5,14,23,32],index=['a','b','c','d'])
print('To add series1 and series2')
print('---------------------------------------------')
print(s1+s2)
print('To add series2 and series3')
print('---------------------------------------------')
print(s2+s3)
print('To add series2 and series3 and filled non matching index with 0')
print('---------------------------------------------')
print(s2.add(s3,fill_value=0))

```
